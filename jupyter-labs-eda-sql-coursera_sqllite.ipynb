## Tasks

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
df = %sql select * from SPACEXTABLE
df = df.DataFrame()
df.head()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
df_cca = %sql select * from SPACEXTABLE where Launch_Site like '%CCA%'
df_cca = df_cca.DataFrame()
df_cca.head()
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
df_payload_mass = %sql SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Customer LIKE '%NASA (CRS)%'
df_payload_mass = df_payload_mass.DataFrame()
df_payload_mass
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
avg_payload_mass = %sql SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Booster_Version LIKE '%F9 v1.1%'
avg_payload_mass = avg_payload_mass.DataFrame()
avg_payload_mass
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
first_launch_date = %sql SELECT MIN(Date) FROM SPACEXTABLE
first_launch_date = first_launch_date.DataFrame()   
first_launch_date
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
booster_version_counts = %sql SELECT Booster_Version, COUNT(*) as count FROM SPACEXTABLE GROUP BY Booster_Version ORDER BY count DESC
booster_version_counts = booster_version_counts.DataFrame() 
booster_version_counts
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
no_of_launches = %sql SELECT COUNT(*), Mission_Outcome FROM SPACEXTABLE GROUP BY Mission_Outcome
no_of_launches = no_of_launches.DataFrame()
no_of_launches
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
max_payload_boosters = %sql \
SELECT DISTINCT Booster_Version \
FROM SPACEXTABLE \
WHERE PAYLOAD_MASS__KG_ = ( \
    SELECT MAX(PAYLOAD_MASS__KG_) \
    FROM SPACEXTABLE \
)

max_payload_boosters = max_payload_boosters.DataFrame()
max_payload_boosters
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
df_2015_failures = %sql \
SELECT substr(Date, 6, 2) AS Month, \
       Landing_Outcome, \
       Booster_Version, \
       Launch_Site \
FROM SPACEXTABLE \
WHERE substr(Date, 1, 4) = '2015' \
AND Landing_Outcome LIKE 'Failure (drone ship)%'

df_2015_failures = df_2015_failures.DataFrame()
df_2015_failures
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
landing_rank = %sql \
SELECT Landing_Outcome, \
       COUNT(*) AS Outcome_Count \
FROM SPACEXTABLE \
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20' \
GROUP BY Landing_Outcome \
ORDER BY Outcome_Count DESC

landing_rank = landing_rank.DataFrame()
landing_rank
```
